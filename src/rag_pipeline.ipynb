```python
import json
import os
```

```python
def load_json_files(folder_path='/content/data'):
  data = []
  for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
      file_path = os.path.join(folder_path, filename)
      with open(file_path, 'r', encoding='utf-8') as f:
        try:
          data.append(json.load(f))
        except json.JSONDecodeError as e:
          print(f"Error decoding JSON in file {filename}: {e}")
  return data
```

```python
from sentence_transformers import SentenceTransformer
```

```python
mpnet_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')
labse_model = SentenceTransformer("sentence-transformers/LaBSE")
stn_model = SentenceTransformer("Yunika/sentence-transformer-nepali")
nem_model = SentenceTransformer("universalml/Nepali_Embedding_Model")
```

```python
import chromadb
```

```python
class MultilingualEmbeddingFunction(chromadb.EmbeddingFunction):
  def __init__(self, embedding_model):
    self.model = embedding_model

  def __call__(self, input: chromadb.Documents)-> chromadb.Embeddings:
    embeddings = self.model.encode(input)
    return embeddings.tolist()
```

```python
from typing import List

def create_collection(documents: List, path: str, name: str, embedding_model):
  client = chromadb.PersistentClient(path=path)
  embedding_function = MultilingualEmbeddingFunction(embedding_model)

  try:
    collection = client.create_collection(name=name, embedding_function=embedding_function)
    k=0
    for i in range(len(documents)):
      for j in range(documents[i]["total_chunks"]):
        k = k+1
        collection.add(documents=documents[i]["chunks"][j]["content"], ids=str(k))
    return collection

  except Exception as e:
    print(f"Error creating collection: {e}")
```

```python
def load_collection(path:str, name:str, embedding_model):
  client = chromadb.PersistentClient(path=path)
  collection = client.get_collection(name=name, embedding_function=MultilingualEmbeddingFunction(embedding_model))
  return collection
```

```python
def get_relevant_chunks(query: str, db, n_results: int=3):
  results = db.query(query_texts=[query], n_results=n_results, include=["documents", "distances", "metadatas"])
  return results
```

```python
def rag_prompt(query: str, relevant_chunks: List[str]):
  context = "\n\n".join([f'''PASSAGE {i+1}:{p.replace("'", "").replace('"', '').strip()}''' for i, p in enumerate(relevant_chunks)])
  return f"""
  You are a helpful AI assistant that answers questions about Nepali government document processes
  (such as citizenship, passport, and birth certificates).

  Follow these rules strictly:

  1. Answer *only* using the provided context.
  2. If the answer is not directly given, you may *infer it indirectly* from related information in the context —
    but only if the inference is *logical and clearly supported* by the context.
    - Note: These are legal/government documents, so you must be extremely careful when inferring answers.
  3. If neither direct nor reasonable indirect inference is possible, reply exactly:
    "मलाई थाहा छैन।"  (I don’t know.)

  *CRITICAL FORMATTING RULES:*
  - Keep answers *concise*
  - Make sure answer is not vague try to resolve vagueness
  - Use bullet points only when necessary
  - Use *simple and clear language*

  *Answer Style Guide:*
  - *Document lists:* Very short bullet points
  - *Procedures:* Minimal step-by-step format
  - *Definitions:* One clear sentence
  - *Requirements:* Minimal list or short sentence

  QUESTION: {query}
  CONTEXT:
  {context}
  ANSWER:
  """

```

```python
def generate_response(db, query:str):
  relevant_chunks = get_relevant_chunks(query, db)['documents'][0]
  # print(f"Relevant Chunks: {relevant_chunks}")
  prompt = rag_prompt(query, relevant_chunks)
  model = genai.GenerativeModel("gemini-2.5-flash")
  try:
    response = model.generate_content(prompt)
    return response.text
  except Exception as e:
    return f"Error generating response: {str(e)}"
```

```python
db_paths = ['/content/db_mpnet', '/content/db_labse', '/content/db_nem', '/content/db_stn']
embedding_models = [mpnet_model, labse_model, nem_model, stn_model]
dbs =[]
for i, db_path in enumerate(db_paths):
  os.makedirs(db_path, exist_ok=True)
  collection_name = 'collection'

  db = create_collection(data, db_path, collection_name, embedding_models[i])
  print("DB created successfully")
  print(f"Collection contains {db.count()} items")
  dbs.append(db)
```

```python
qa_pairs = load_json_files(folder_path='/content/qa')
print(qa_pairs)
```

```python
print(qa_pairs[0]["citizenship_qa"][0]['question'])
```

```python
questions = []
answers = []

for qa_pair in qa_pairs[0]["citizenship_qa"]:
  questions.append(qa_pair['question'])
  answers.append(qa_pair['answer'])
```

```python
import time

filename = ["mpnet_response.json", "labse_response.json", "nem_response.json", "stn_response.json"]
for i in range(len(dbs)):
  qa_response = []
  for j in range(len(questions)):
    output = generate_response(dbs[i], questions[j])
    print(f"Answered {j+1} out of {len(questions)}")
    output_response_json = {
        'question': questions[j],
        'answer': output
    }
    qa_response.append(output_response_json)
    time.sleep(5)

  with open(filename[i], 'w', encoding="utf-8") as f:
    json.dump(qa_response, f, ensure_ascii=False, indent=4)
  print(f"Saved {filename[i]}")
  print("\n" + "="*50 + "\n")
```

```python
import matplotlib.pyplot as plt

idk_response = [12, 8, 6, 9]
labels = ['mpnet', 'labse', 'nem', 'stn']

plt.bar(labels, idk_response)
plt.title("I don't Know: Response Distribution")
plt.xlabel('Embedding Models')
plt.ylabel('Count out of 20')
plt.ylim(0, 20)
plt.show()
```

The evaluation used the following scoring criteria for each question:

**1.0 (Fully Correct)**: The answer is factual, complete, and concise, matching the ground truth.

**0.5 (Partially Correct)**: The answer contains important facts but misses key details or is incomplete.

**0.0 (Incorrect / No Answer)**: The answer is factually wrong or the model responded with "I don't know" (indicating a failure to retrieve or generate).
